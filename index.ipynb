```python
# Your code here - prepare data for modeling
# log('Garage Area', 'GrLivArea', 'TotalBsmsft', '1stFlrSF', 'GarageYrBlt') are used
```

```python
# Log transform features and SalePrice
ames_log = ames[['GarageArea', 'GarageYrBlt', 'GrLivArea', 'YearBuilt', '1stFlrSF', 'TotalBsmtSF', 'SalePrice']].copy()
ames_log = ames_log[ames_log > 0]  # Ensure all values are positive for log transformation
ames_log = np.log(ames_log.dropna())

# Prepare the X dataframe and y series
x_log = ames_log[['GarageArea', 'GrLivArea', 'TotalBsmtSF']]
y = ames_log['SalePrice']
```

### Modeling

Now build a StatsModels OLS model with a log-transformed target as well as log-transformed features.

```python
# Your code here - build a model
import statsmodels.api as sm


log_model = sm.OLS(y, sm.add_constant(x_log))
log_model_results = log_model.fit()

print(log_model_results.summary())



```

```python
for column in log_model_results.params.index[1:]:
    param = (np.exp(log_model_results.params[column] * np.log(1.01)) -1  )* 100
    print(param)
```
